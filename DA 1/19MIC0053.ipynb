```python
data['Pregnancies']=data['Pregnancies'].replace(np.NaN,0) 
data['Glucose']=data['Glucose'].replace(np.NaN,0) 
data['BloodPressure']=data['BloodPressure'].replace(np.NaN,0) 
data['SkinThickness']=data['SkinThickness'].replace(np.NaN,0) 
data['Insulin']=data['Insulin'].replace(np.NaN,0) 
data['BMI']=data['BMI'].replace(np.NaN,0)
data[:10]
```

```python
df=pd.read_csv("diabetes.csv") 
m1=df['Pregnancies'].mean()
m2=df['Glucose'].mean() 
m3=df['BloodPressure'].mean() 
m4=df['SkinThickness'].mean()
m5=df['Insulin'].mean() 
m6=df['BMI'].mean()
df['Pregnancies']=df['Pregnancies'].fillna(m1) 
df['Glucose']=df['Glucose'].fillna(m2) 
df['BloodPressure']=df['BloodPressure'].fillna(m3) 
df['SkinThickness']=df['SkinThickness'].fillna(m4) 
df['Insulin']=df['Insulin'].fillna(m5) 
df['BMI']=df['BMI'].fillna(m6)
df[:10]
```

```python
df=pd.read_csv("Diabetes.csv") 
me1=df['Pregnancies'].median() 
me2=df['Glucose'].median() 
me3=df['BloodPressure'].median() 
me4=df['SkinThickness'].median() 
me5=df['Insulin'].median()
me6=df['BMI'].median()
df['Pregnancies']=df['Pregnancies'].fillna(me1) 
df['Glucose']=df['Glucose'].fillna(me2) 
df['BloodPressure']=df['BloodPressure'].fillna(me3) 
df['SkinThickness']=df['SkinThickness'].fillna(me4) 
df['Insulin']=df['Insulin'].fillna(me5) 
df['BMI']=df['BMI'].fillna(me6)
df[:10]
```
